# Feature Importance Analysis 

```python
## Top 10 features based on feature importance analysis 
## feature importance analysis done with random forest classifier. 
```

# Split data into training and Test Data

```python
X = ds.iloc[:, :-1].values
y = ds.iloc[:, -1].values
```

```python
Xtrain, Xtest, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 43)
```

```python
print(Xtrain)
```

```python
print(Xtest)
```

```python
print(y_train)
```

```python
print(y_test)
```

# Train Model for Feature Importance 

```python
RF_classifier_FI = RandomForestClassifier(n_estimators = 100, random_state =42)
RF_classifier_FI.fit(Xtrain, y_train)
```

```python
important_features = RF_classifier_FI.feature_importances_
```

```python
important_features.sort()
```

```python
top_10 = important_features[important_features> 0.03011984790603075]
```

```python
new_ds = ds.drop(['Class'], axis=1)
```

```python
new_ds.shape
```

```python
feature_importance_ds = pd.DataFrame()
```

```python
feature_importance_ds = new_ds[['modules.nmodules', 'svcscan.nservices', 'svcscan.kernel_drivers', 'svcscan.fs_drivers', 
                               'svcscan.process_services', 'svcscan.shared_process_services', 'svcscan.nactive',
                               'callbacks.ncallbacks', 'callbacks.nanonymous', 'callbacks.ngeneric']]
```

# Visualize Feature Importance 

```python
## Visualizing the Feature Importances. 

importances = pd.Series(top_10, index = feature_importance_ds.columns).sort_values(ascending=False)
plt.subplots(figsize=(10, 10))
sns.barplot(x=importances, y = importances.index, palette='coolwarm')
plt.xlabel('Feature Importance Score', fontsize=12)
plt.ylabel('Features', fontsize=10)
plt.title("Visualizing Feature Importances", fontsize=15)
```
